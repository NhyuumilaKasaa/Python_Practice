```python
# 72. Use a generator to create an infinite sequence of odd numbers.

def odd():
  n=1
  while True:
    yield n # returns a value and pauses the function's execution. next() resumes it again.
    n=n+2

oddnum=odd()
for i in range(1:10):
  print(next(oddnum)) # The next() resumes the function again and again.
```

```python
# 90. Write a program to validate an email address using regex.

import re

def val(email):
  regex=r'^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'
  if re.match(regex,email):
    return True
  else:
    return False

email1="hello@gmail.com"
email2="invalid_email"

if val(email1):
  print("Valid email.")
else:
  print("Invalid email.")

if val(email2):
  print("Valid email.")
else:
  print("Invalid email.")
```
