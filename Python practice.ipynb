```python
#Fibonacci series
n=6
f=0
s=1
i=0
while(i<n):
  if i==1:
    next=i
  else:
    next=f+s
    f=s
    s=next
  i=i+1
  print("\t",next)
```

```python
#HCF and LCM
a=2
b=4
def func(x,y):
  if x>y:
    small=y
  else:
    small=x
  for i in range(1,small+1):
    if x%i==0 and y%i==0:
      hcf=i
  return hcf
hcf=func(a,b)
lcm=(a*b)/hcf
print("HCF is:",hcf)
print("LCM is:",lcm)
```
